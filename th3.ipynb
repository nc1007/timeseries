```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file
df = pd.read_csv('/content/Gia SMP va SMPcap 2021(Giá thị trường SMP).csv',encoding="latin-1",sep = ';')
df.head()
```

```python
# Chọn cột 2, 3 và 4
X = df.iloc[:, [8, 9, 10]].values
```

```python
# Trực quan hóa phân phối của cột 2, 3 và 4
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(data=df, x=df.columns[8], kde=True, color='blue')
plt.title('Distribution of ' + df.columns[8])

plt.subplot(1, 3, 2)
sns.histplot(data=df, x=df.columns[9], kde=True, color='green')
plt.title('Distribution of ' + df.columns[9])

plt.subplot(1, 3, 3)
sns.histplot(data=df, x=df.columns[10], kde=True, color='red')
plt.title('Distribution of ' + df.columns[10])

plt.tight_layout()
plt.show()
```

```python
# Trực quan hóa mối quan hệ giữa các cột
sns.pairplot(df.iloc[:, [8, 9, 10]])
plt.show()

# Nếu muốn hiển thị các mối quan hệ trong cùng một biểu đồ với scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df.iloc[:, 8], y=df.iloc[:, 9], hue=df.iloc[:, 10], palette='viridis')
plt.title('Relationship between ' + df.columns[8] + ' and ' + df.columns[9] + ' with hue of ' + df.columns[10])
plt.xlabel(df.columns[8])
plt.ylabel(df.columns[9])
plt.show()
```

```python
from filterpy.kalman import KalmanFilter
import numpy as np

# Function to apply Kalman Filter
def apply_kalman_filter(data):
    kf = KalmanFilter(dim_x=2, dim_z=1)
    kf.x = np.array([0., 0.])       # initial state (location and velocity)
    kf.F = np.array([[1., 1.], [0., 1.]]) # state transition matrix
    kf.H = np.array([[1., 0.]])     # Measurement function
    kf.P *= 1000.                   # covariance matrix
    kf.R = 5                        # state uncertainty
    kf.Q = np.array([[0.1, 0.1], [0.1, 0.1]]) # process uncertainty

    filtered_data = []
    for z in data:
        kf.predict()
        kf.update(z)
        filtered_data.append(kf.x[0])
    return filtered_data

# Áp dụng Kalman Filter cho cột thứ 2 (Feature_2)
filtered_feature_2 = apply_kalman_filter(X[:, 0])

# So sánh dữ liệu gốc và dữ liệu đã lọc
plt.figure(figsize=(10, 6))
plt.plot(X[:, 0], label='Original ' + df.columns[8])
plt.plot(filtered_feature_2, label='Filtered ' + df.columns[8], linestyle='dashed')
plt.title('Kalman Filter on ' + df.columns[14])
plt.xlabel('Index')
plt.ylabel(df.columns[8])
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

# Tạo một DataFrame mới với cột 2, cột 3 đã được lọc, và cột 4
df_filtered = pd.DataFrame({
    df.columns[8]: X[:, 0],
    df.columns[9]: filtered_feature_3,
    df.columns[10]: X[:, 2]
})

# Sử dụng k-means clustering với 3 cụm
kmeans = KMeans(n_clusters=3, random_state=42)
df_filtered['Cluster'] = kmeans.fit_predict(df_filtered)

# Trực quan hóa kết quả cụm
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_filtered[df.columns[9]], y=df_filtered[df.columns[10]], hue=df_filtered['Cluster'], palette='viridis')
plt.title('K-means Clustering on Filtered Data')
plt.xlabel(df.columns[8])
plt.ylabel(df.columns[9])
plt.show()
```
